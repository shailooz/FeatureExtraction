```python
from tensorflow import keras
import tensorflow as tf
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,LabelBinarizer
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Input,Dense,Concatenate

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :]  # we only take the first two features.
y = iris.target
y = tf.keras.utils.to_categorical(y, num_classes=3)
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

tf.random.set_seed(42)


model = Sequential([
 Input((4,)),
 Dense(100, activation='relu'),
 Dense(50, activation='relu'),
 Dense(3, activation='softmax')
])
'''
model = Sequential()
model.add(Dense(100, activation='relu', input_shape = (4,)))
model.add(Dense(50, activation='relu'))
model.add(Dense(3, activation='softmax'))
input = Input(shape=(4,))
x = Dense(100, activation='relu')(input)
y = Concatenate()([input,x])
x = Dense(50, activation='relu')(y)
output = Dense(3, activation='softmax')(x)
model = Model(input, output )
'''

model.compile(optimizer='sgd',loss="categorical_crossentropy",metrics=['accuracy'])

model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model, show_shapes=True)
```

```python

model.fit(X_train,y_train,epochs=100,batch_size=16)#no of iterations-epochs,
model.evaluate(X_test,y_test)
```

```python
y_pred  = np.argmax(model(X_test), axis =-1)
print(y_pred)

y_true =  np.argmax(y_test, axis =-1)
print(y_true)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_true, y_pred)
print(cm)
cr = classification_report(y_true, y_pred)
print(cr)
```

```python
import seaborn as sns
sns.heatmap(cm,cmap='Greens', annot=True,xticklabels=iris.target_names,yticklabels=iris.target_names)
```

# **Image Classfication**

```python
import zipfile

with zipfile.ZipFile('/content/dataset.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
# prompt: /content/dataset in the given folder i kept my data in two subfolders load all images and resize to 300,300  in X and labels in Y

import os
from PIL import Image
import numpy as np
import cv2

X = []
Y = []

# Assuming the extracted dataset is in a directory named 'dataset'
dataset_path = '/content/dataset'
label={"lemon":1,"melon":0}

for label_folder in os.scandir(dataset_path):
  for image_file in os.scandir(label_folder.path):
    img = cv2.imread(image_file.path)
    X.append(cv2.mean(img)[:3])
    Y.append(label[label_folder.name]) # Use the folder name as the label

X = np.array(X)
Y = np.array(Y)

print(f"Image data shape: {X.shape}")
print(f"Labels shape: {Y.shape}")
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

tf.random.set_seed(42)


model = Sequential([
 Input((3,)),
 Dense(100, activation='relu'),
 Dense(50, activation='relu'),
 Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='sgd',loss="binary_crossentropy",metrics=['accuracy'])
model.summary()
model.fit(X_train,y_train,epochs=10,batch_size=2)#no of iterations-epochs,
model.evaluate(X_test,y_test)
```

```python
y_pred  = np.array(model(X_test)>0.5, dtype=int).ravel()
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)
cr = classification_report(y_test, y_pred)
print(cr)
```
